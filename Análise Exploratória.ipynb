```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import re
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from datetime import datetime
```

```python
#Leitura dos dados
nike_data = pd.read_csv(r"C:\Users\mates\Desktop\nike_data.csv")
```

#### Número de exemplares

```python
num_exemplares = len(nike_data)
print("Número de exemplares:", num_exemplares)
```

#### Valor máximo e mínimo

```python
valor_maximo = nike_data['preco'].max()
valor_minimo = nike_data['preco'].min()
print("Valor máximo:", valor_maximo)
print("Valor mínimo:", valor_minimo)
```

#### Variância e desvio padrão

```python
variancia = nike_data['preco'].var()
desvio_padrao = nike_data['preco'].std()
print("Variância:", variancia)
print("Desvio padrão:", desvio_padrao)
```

#### Distribuição

```python
plt.hist(nike_data['preco'], bins=25, color='white', edgecolor='black')
plt.title('Distribuição dos preços dos produtos')
plt.xlabel('Preço')
plt.ylabel('Frequência')
plt.show()
```

#### Quantidade de Nulos

```python
quantidade_nulos = nike_data.isna().sum()
print("Quantidade de Nulos:")
print(quantidade_nulos)
```

#### Existência de outliers

```python
sns.boxplot(x=nike_data['preco'])
plt.title('Boxplot do preço')
plt.xlabel('Preço')
plt.show()
```
